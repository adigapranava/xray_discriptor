```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import xml.etree.ElementTree as ET
import os
import cv2
import tensorflow as tf
from wordcloud import WordCloud
import re
from collections import defaultdict
import itertools
from collections import Counter
```

```python
import xml.etree.ElementTree as ET
img = []
img_impression = []
img_finding = []
reqd =[]
```

```python
directory = "/work/reports/ecgen-radiology"
```

![Picture title](image-20220617-064150.png)

```python
from tqdm import tqdm
for filename in tqdm(os.listdir(directory)):
    if filename.endswith(".xml"):
        f = directory + '/' + filename
        tree = ET.parse(f)
        root = tree.getroot()
        obj = {}
        for child in root:
            if child.tag == 'MedlineCitation':
                for attr in child:
                    if attr.tag == 'Article':
                        for i in attr:
                            if i.tag == 'Abstract':
                                for name in i:
                                    if name.get('Label') == 'FINDINGS':
                                        finding=name.text
                                        obj["desc"] = finding
        obj["imgs"] = []
        for p_image in root.findall('parentImage'):
            img.append(p_image.get('id'))
            img_finding.append(finding)
            obj["imgs"].append(p_image.get('id'))
        reqd.append(obj)
```

```python
import contractions
text = "She'd like to know how I'd do that! i couldn't come here"
contractions.fix(text)
```

```python
def text_cleaner(text):
    # Lower Casing the Data
    text = text.lower()
    # remove "XXXX"
    text = re.sub("(x){2,}", "", text)
    # Removing Numbers
    text = ''.join([i for i in text if not i.isdigit()])
    # Removing Contractions
    text = contractions.fix(text)
    # Remove Special Char
    text = re.sub(r"[,.;@#?!&$<>[\]{}()~^*-:'\"\\]+\ *", " ", text)
    # Removing Extra Space
    text = " ".join(text.split())
    # Replacing the Repetitions of dots(.)
    text = re.sub("(\.){2,}", "", text)
    # Replacing the Repetitions of Punctuations
    text = re.sub("(\!){1,}", "", text)
    # separete fullstop from word
    text = re.sub('\.', ' .', text)
    temp = []
    # Removes words less than 2 characters except no and ct
    for word in text.split():
        if  len(word) <=2 and word != 'no' and word != 'ct':
            continue
        else:
            temp.append(word)
    text = " ".join(temp)
    return text
```

```python
def train_test_split(data, test_size=0.4):
    size = len(data)
    train_index = int(size*(1-test_size))
    persons_train = data[:train_index]
    persons_test = data[train_index: train_index+int(size*(test_size/2))]
    persons_cv = data[train_index+int(size*(test_size/2)):]
    return persons_train, persons_cv, persons_test
```

```python
df_train, df_cv, df_test = train_test_split(data_frame,0.3)
df_train.to_csv("df_train.csv")
df_test.to_csv("df_test.csv")
df_cv.to_csv("df_cv.csv")
```

```python
def load_image(file):
    img = tf.io.read_file(file)
    img = tf.image.decode_png(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return img
```
